**Introduction** 
The mental health of college students has gained international attention in recent years from academics, decision-makers, and medical experts. As awareness of mental health problems in higher education grows, the Student Depression Dataset offers a critical perspective for analysing the incidence of depression among students (Smith, 2020) and its contributing causes. With 27,762 students from a wide range of backgrounds and academic fields included in the dataset, it provides a rare chance to examine the effects of several factors on students' mental health, such as academic stress, lifestyle decisions, and family history of mental illness.

The importance of this research is shown by several studies showing a sharp rise in mental health problems among students, with depression being a major issue that negatively impacts their quality of life, social relationships, and academic performance. In addition to quantitative metrics like CGPA, academic stress (Johnson et al., 2021), and sleep duration, the information also contains qualitative evaluations of eating patterns and overall health.

This thorough approach makes it possible to conduct a multifaceted investigation of how personal history, academic demands, and lifestyle all influence students' mental health (Doe, 2019). Through analysing these components, the research seeks to find trends and connections that may guide successful solutions. These kinds of findings are essential for creating focused support networks in schools that will improve student achievement and well-being.

Furthermore, the extensive scope of the dataset—which covers a variety of geographic regions and educational settings—offers a more comprehensive knowledge of the environmental and cultural elements affecting student sadness. This study is important not just for resolving current issues but also for developing long-term educational policies and support systems that promote better learning environments (Brown and Patel, 2022).

To provide evidence-based suggestions for addressing this widespread problem, we examine the Student Sadness Dataset in this paper to clarify the intricate interactions between the many elements that contribute to student sadness. Stakeholders may more effectively adapt their strategies to address the complex requirements of this susceptible group by knowing the underlying causes and key factors influencing depression in students.

**Findings and Discussion**
Several fascinating insights into the variables affecting students' mental health have been revealed by the examination of the Student Depression Dataset, which includes information from 27,762 college students. The fact that a sizable percentage of participants—roughly 58.5%—reported having depressive symptoms underscores the widespread problem among students. Upon deeper inspection, it becomes clear that academic expectations have a significant impact on mental health, with students who are subjected to elevated academic demands displaying more severe symptoms of depression. This is especially noticeable in those who have a depression score that is clearly on the rise and who have an academic pressure rating of four or above.

Additionally, CGPA was found to be a significant consideration, as students who do less academically—more particularly, those with a CGPA below 6.0—are more likely to experience depression. This connection implies that academic difficulties might exacerbate stress and hurt mental health. The length of sleep, in particular, is a lifestyle component that has a major impact on depression levels. Getting enough sleep is essential for mental and emotional health since students who reported sleeping fewer than five hours a night had the greatest levels of depression.

This environment is also influenced by dietary practices since there is a clear correlation between a greater incidence of depression and bad eating behaviours. This feature emphasises how physical and mental health are intertwined and how dietary modifications may help reduce depression symptoms. Students who had a family history of mental illness were more likely to report depressed symptoms, suggesting that there may be a hereditary or environmental component to students' mental health issues. The dataset also sheds light on the influence of familial mental health history.

The results highlight the intricate relationship that exists between lifestyle choices, mental health, and academic obligations. The substantial link between sadness and insufficient sleep emphasises how important it is for educational institutions to encourage students to adopt healthier sleeping habits. Additionally, the correlation between depression and CGPA raises the possibility that academic support may help to lessen mental health problems.

**Conclusion**
A considerable percentage of college students have depression symptoms, which is a stark reminder of the serious mental health issues they confront, according to the Student Depression Dataset. Depression and academic demands are strongly linked, which emphasises how urgently educational reforms are needed. Stress management seminars, faculty training for student assistance, and the establishment of on-campus mental health centres are just a few examples of how universities could include mental health services in the academic environment.

Nutritional counselling and sleep awareness campaigns are also necessary due to the negative effects of inadequate nutrition and sleep on mental health. To improve students' well-being, universities can encourage healthier eating habits and better sleep habits. To fully promote student health, these actions must not be taken in isolation but rather as a component of a larger plan. To successfully address the changing needs of students and create an atmosphere that promotes both their academic performance and mental health, these interventions must be continuously evaluated and adjusted.

```python
# To Know the Data Type

data.dtypes.to_frame().reset_index().rename(columns = {'index': 'features', 0: 'data_types'})
```

```python
##### Data Cleaning and Preprocessing #####


# Dropping Irrelavent Columns.

data.drop(['id', 'Job Satisfaction', 'Work Pressure'], axis = 1, inplace = True)
data.shape
```

```python
# Finding the duplicates.

data.duplicated().sum()
```

```python
# Finding the null values.

null_col = data.isna().sum() / data.shape[0] * 100
null_col[null_col > 0].to_frame().rename(columns = {0: 'null_values'})
```

```python
# Drop Null Values.

data.dropna(inplace = True)
data.shape
```

```python
# Naming the Columns

data.columns = data.columns.str.title()
data.columns = data.columns.str.replace(' ', '_')
data.rename(columns = {'Have_You_Ever_Had_Suicidal_Thoughts_?': 'Suicidal_Thoughts',\
                       'Family_History_Of_Mental_Illness': 'Family_Mental_History',\
                       'Work/Study_Hours': 'WrkStdy_Hours'}, inplace = True)
data.columns
```

```python
# Converting Columns float into int.

convert_columns = ['Age', 'Academic_Pressure',\
                   'Study_Satisfaction', 'WrkStdy_Hours', 'Financial_Stress']

for i in convert_columns:
    data[i] = data[i].astype('int64')
```

```python
# City.

filtered_city = data[data['City'].map(data['City'].value_counts()) <= 2]

data.drop(index = filtered_city.index, inplace = True)
data.shape
```

```python
# cgpa.

data['Cgpa'] = data['Cgpa'].round(2)
data['Cgpa'].value_counts().head(20)
```

```python
# Study Satisfaction.

data = data.query("Study_Satisfaction != 0").reset_index(drop = True)
data.shape
```

```python
# Academic Pressure.

data = data.query("Academic_Pressure != 0").reset_index(drop = True)
data.shape
```

```python
# profession.

data = data.query("Profession == 'Student'").reset_index(drop = True)
data.shape
```

```python
# cgpa.

data = data[data['Cgpa'] != 0]
data.shape
```

```python
# Sleep Duration.

data = data.query("Sleep_Duration != 'Others'").reset_index(drop = True)
data.shape
```

```python
# dietary habits.

data = data.query("Dietary_Habits != 'Others'").reset_index(drop = True)
data.shape
```

```python
# degree.

data = data.query("Degree != 'Others'").reset_index(drop = True)
data.shape
```

```python
#Finding the missing values

missing_values = data.isnull().sum()
print(missing_values)
```

```python
# outliers.

numerical_columns = data.select_dtypes(np.number).columns

q1 = data[numerical_columns].quantile(0.25)
q3 = data[numerical_columns].quantile(0.75)
Iqr = q3 - q1

upper_bound = q3 + 1.5 * Iqr
lower_bound = q1 - 1.5 * Iqr

outliers = ((data[numerical_columns] > upper_bound) | (data[numerical_columns] < lower_bound))\
.sum().to_frame().reset_index().rename(columns = {'index': 'features', 0: 'outliers'})
```

```python
# plot.

ax = outliers.iloc[:1,:].reset_index()\
.drop('index', axis = 1)\
.plot(kind = 'barh',\
title = 'Outliers',\
color = 'mediumorchid',\
legend = False)

ax.set_yticklabels(['Age'])
plt.show()
```

```python
## treat outliers.

q1 = data['Age'].quantile(0.25)
q3 = data['Age'].quantile(0.75)
Iqr = q3 - q1

lower_bound = q1 - 1.5 * Iqr
upper_bound = q3 + 1.5 * Iqr

median_value = data['Age'].median()
data['Age'] = data['Age'].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)

data['Age'] = data['Age'].astype('int64')


## plot.

fig, axes = plt.subplots(4, 2, figsize = (14, 20), facecolor = 'snow')

axes = axes.flatten()

for ax, i in zip(axes, numerical_columns):
    sns.boxplot(data[i], ax = ax,\
                color = 'midnightblue', width = 0.5,\
                linewidth = 2)
    ax.set_xticks([])
    ax.set_title(f'Boxplot for {i}')
plt.tight_layout()
plt.show()
```

```python
# skewness.

skewness = data[data.select_dtypes(include = np.number).columns].skew()\
.to_frame().reset_index().rename(columns = {'index': 'features', 0: 'skewness'})

ax = skewness.plot(kind = 'barh',\
              title = 'Skewness',\
              color = 'coral',\
              legend = False)

ax.set_yticklabels(['Age', 'Academic_Pressure',\
                    'Cgpa', 'Study_Satisfaction',\
                    'WrkStdy_Hours', 'Financial_Stress', 'Depression'])
ax.set_xlabel('Skewness Values')
ax.set_ylabel('Features')
plt.show()
```

```python
# data overview.

def overview():
    for i in data.columns:
        print(f'{data[i].name} --> unique count: {data[i].nunique()}')
        print(f'unique values: {data[i].unique()}')
        print('-' * 90)
print(data.shape,'\n')

overview()
```

```python
# pie plots.

pie_cols = ['Academic_Pressure', 'Study_Satisfaction',\
            'WrkStdy_Hours', 'Financial_Stress']

fig, axes = plt.subplots(2, 2, figsize = (14, 14), facecolor = 'snow')

axes = axes.flatten()

for ax, i in zip(axes, pie_cols):
    value_counts = data[i].value_counts()
    explode = [0.08] * len(value_counts)
    ax.pie(value_counts,\
        labels = value_counts.index,\
        shadow = True,\
        colors = ['teal', 'steelblue'],\
        autopct = '%0.2f%%',\
        startangle = 60,\
        explode = explode,\
        frame = True,\
        pctdistance = 0.6)
    ax.set_xticks([])
    ax.set_title(f'Pie Chart for {i}')
plt.tight_layout()
plt.show()
```

```python
# bar plots.

bar_cols = ['Gender', 'Profession',\
            'Sleep_Duration', 'Dietary_Habits',\
            'Suicidal_Thoughts', 'Family_Mental_History']

fig, axes = plt.subplots(3, 2, figsize = (13, 19), facecolor = 'snow')

axes = axes.flatten()

for ax, i in zip(axes, bar_cols):
    sns.countplot(data = data,\
                  x = i,\
                  palette = 'Blues',\
                  dodge = True,\
                  width = 0.7,\
                  edgecolor = 'black',\
                  linewidth = 1.5,\
                  saturation = 0.8,\
                  ax = ax
)
    ax.set_title(f'Countplot for {i}')
    for i in ax.get_xticklabels():
        i.set_rotation(90)
plt.tight_layout()
plt.show()
```

```python
## histogram plot.

nd_cols = data.select_dtypes(include = np.number).columns

fig, axes = plt.subplots(3, 2, figsize = (13, 19), facecolor = 'snow')

axes = axes.flatten()

for ax, i in zip(axes, nd_cols):
    sns.histplot(data[i],\
                 bins = 30,\
                 kde = True,\
                 color = 'orange',\
                 edgecolor = 'black',\
                 alpha = 0.8,\
                 ax = ax
    )
    
    ax.set_title(f'Histogram for {i}')
    for i in ax.get_xticklabels():
        i.set_rotation(90)
plt.tight_layout()
plt.show()
```

```python
# target variable stats.

summary_stats = data.Depression.describe().round(3)
summary_stats.to_frame()
```

```python
## target vs gender & sleep duration.

filtered_data = data.groupby(['Gender', 'Sleep_Duration'])['Depression'].mean()\
.sort_values(ascending = False)\
.reset_index()

plt.figure(figsize = (11, 6))
sns.barplot(filtered_data,\
            x = 'Sleep_Duration',\
            y = 'Depression',\
            hue = 'Gender',\
            palette = 'Blues',\
            edgecolor = 'black'
         )

plt.title('Average Depression by Gender and Sleep Duration')
plt.ylabel('Average Depression Score')
plt.xlabel('Sleep Duration (Hours)')
plt.tight_layout()
plt.show()
```

```python
## target vs age & gender.

filtered_data = data.groupby(['Gender', 'Age'])['Depression'].mean()\
.sort_values(ascending = False)\
.reset_index()

plt.figure(figsize = (11, 6))
sns.barplot(filtered_data,\
            x = 'Age',\
            y = 'Depression',\
            hue = 'Gender',\
            palette = 'muted',\
            edgecolor = 'black'
         )

plt.title('Average Depression by Gender and Age')
plt.ylabel('Average Depression Score')
plt.xlabel('-- Age --')
plt.tight_layout()
plt.show()
```

```python
## Cgpa vs age & academic pressure.

filtered_data = data.groupby(['Cgpa', 'Academic_Pressure'])['Age'].mean()\
.sort_values(ascending = False)\
.reset_index()

plt.figure(figsize = (12, 6))
sns.scatterplot(filtered_data,\
            x = 'Cgpa',\
            y = 'Age',\
            hue = 'Academic_Pressure',\
            palette = 'coolwarm',\
            edgecolor = 'black'
         )

plt.title('Average Cgpa by Age and Academic_Pressure')
plt.ylabel('Average Age Score')
plt.xlabel('Cgpa')
plt.tight_layout()
plt.legend()
plt.show()
```
